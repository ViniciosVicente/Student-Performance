```python
# Passo 1: Importar as bibliotecas
import pandas as pd

df = pd.read_csv("data.csv")
display(df)
```

```python
# Passo 2: Preparar a base de dados para inteligência artificial

# Colunas que irei utilizar para prever 
from sklearn.preprocessing import StandardScaler
x = df.drop(columns= "Grades")
y = df["Grades"]

x_random = x.copy()

scaler = StandardScaler()
x_sc= scaler.fit_transform(x)

# Separar dados em dados de treino e teste
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x_sc, y, test_size=0.20)
x_treino, x_teste, y_treino, y_teste = train_test_split(x_random, y, test_size=0.20)

display(df.info())

```

```python
# Passo 3 : Treinar a Inteligência Artificial 

# RandomForest
from sklearn.ensemble import RandomForestClassifier
mod_arvoreDecision = RandomForestClassifier()
mod_arvoreDecision.fit(x_treino, y_treino)

#KNN
from sklearn.neighbors import KNeighborsClassifier
mod_knn = KNeighborsClassifier()
mod_knn.fit(x_train, y_train)

#TensorFlow

from tensorflow import keras
from tensorflow.keras import layers


modelo = keras.Sequential([
    keras.layers.Dense(64, activation='relu', input_shape=(4,)),
    keras.layers.Dense(32, activation='relu'),
    keras.layers.Dense(1)
])

modelo.compile(optimizer='adam', loss='mean_squared_error', metrics=['mean_absolute_error'])  
modelo.fit(x_train, y_train, epochs=100, batch_size=50, validation_split=0.2)
```

```python
# Passo 4: Escolher o melhor modelo

previsoes_random = mod_arvoreDecision.predict(x_teste)
previsoes_knn = mod_knn.predict(x_test)
previsoes_tensor = modelo.predict(x_test)

from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

# MAE (Mean Absolute Error): quanto menor, melhor – mostra o erro médio absoluto.
#MSE (Mean Squared Error): também quanto menor, melhor – penaliza erros maiores.
#R² (Coeficiente de Determinação): quanto mais próximo de 1, melhor – mostra o quanto o modelo explica os dados.

# Avaliação
print("Random Forest")
print("MAE:", mean_absolute_error(y_teste, previsoes_random))
print("MSE:", mean_squared_error(y_teste, previsoes_random))
print("R²:", r2_score(y_teste, previsoes_random))

print("\nKNN")
print("MAE:", mean_absolute_error(y_test, previsoes_knn))
print("MSE:", mean_squared_error(y_test, previsoes_knn))
print("R²:", r2_score(y_test, previsoes_knn))

print("\nTensorFlow")
print("MAE:", mean_absolute_error(y_test, previsoes_tensor))
print("MSE:", mean_squared_error(y_test, previsoes_tensor))
print("R²:", r2_score(y_test, previsoes_tensor))
```

```python
# Com o resultado, percebe-se que é mais viável utilizar o RandomForest
```
